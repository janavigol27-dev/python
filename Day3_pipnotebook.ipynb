2. Vector - Row vector 
          - Column Vector

=> Row Vector :

```python
a1=np.array([[1,2,3],[4,5,6]])      #Shape (2,3)
```

```python
a2=np.array([10,20,30])            #Shape(3,)
```

```python
print(a1+a2)
```

```python
#a1.shape==(2,3)
#a2.shape==(3,)     =>broadcast to (2,3)        =>Addee to each row
```

=> column Vector :

```python
a1=np.array([[1,2,3],[4,5,6]])
```

```python
#a2=np.array([[10],[20]])
#a2
a2=np.array([10,20]).reshape(-1,1)
```

```python
print(a1+a2)
```

```python
a2.shape=(2,1)      # => added down each row
```

3.Matrix Broadcasting

```python
a1=np.array([[1],[2]])
a2=np.array([[10,20,30]])
```

```python
print(a1)
print()
print(a2)
print()
```

```python
a3=a1+a2
```

```python
print(a3)
```

```python
#a1 => (2,1) => expended across column.
#a2 => (1,3) => expended across rows.
#Result shape => (2,3)
```
