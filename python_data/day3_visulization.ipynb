1.Bar - plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
```

```python
df=sns.load_dataset("titanic")
```

bar plot

```python
plt.figure(figsize=(8,5))
sns.barplot(x="class",y="fare",data=df,estimator=np.sum,errorbar=None)
plt.xlabel("passenger bar")
plt.ylabel("total")
plt.title("total fare by class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x="class",
    y="fare",
    data=df,
    estimator=np.sum,
    errorbar=None
)
plt.show()
```

```python
# real world style dataset
```

```python
df=pd.DataFrame({
    "Category" : ["Laptop","Laptop","Laptop","Phone","Phone","Tablet","Tablet","Tablet"],
    "Rating" :[4.5,4.0,4.2,4.8,4.6,4.1,4.0,3.9]
})

# bar plot
sns.barplot(x="Category",y="Rating",data=df,errorbar=None,palette="deep",hue="Category")
plt.xlabel("product category")
plt.ylabel("Avg rating")
plt.title("Average Customer Rating by product Category")
plt.show()
```

2.line plot - flights dataset

```python
flights=sns.load_dataset("flights")
line_data=flights.groupby("year")["passengers"].sum().reset_index()
```

```python
# line data
```

```python
# plt.Figure(figsize=(10,5))
# sns.lineplot(data=line_data,x="year",y="passenger",marker="o")
# plt.title("Total passengers per year")
# plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=line_data, x="year", y="passengers", marker="o")
plt.title("Total passengers per year")
plt.show()
```

4.pie chart -titanic dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
# tips = sns.load_dataset("tips")

# # ---------------- Scatter + Regression Chart ----------------
# plt.figure(figsize=(8,5))
# sns.regplot(data=tips, x="total_bill", y="tip", scatter=True, color="green")
# plt.title("Total bill vs tip with regression line")
# plt.show()

# ---------------- Pie Chart ----------------
# Pie chart: Total bill percentage by day
pie_data = tips.groupby("day")["total_bill"].sum()

plt.figure(figsize=(7,7))
plt.pie(
    pie_data,
    labels=pie_data.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Total Bill Distribution by Day (Pie Chart)")
plt.show()
```

6.count plot - titanic dataset

7.heat map - correlation in titanic dataset

```python
df=sns.load_dataset("tips")
corr=df.corr(numeric_only=True)
# corr
plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True,linewidth=0.5,cmap="YlOrRd")
plt.title("correlation heatmap titanic,")
plt.show()
```
